```python
def tokenizer(stop_words,s):
    from nltk.tokenize import WhitespaceTokenizer
    tok = s.split(" ")
    # print stop_words
    # remove stop words
    result = ""
    for w in tok:
        if w.strip() not in stop_words and w.strip() != "":
            result += w.strip()
            result += " "
            
    return result
```

```python
# jieba
import os
import jieba
import chardet
import jieba.analyse
import jieba.posseg as pseg

# load text files
import os
import chardet


# The apart text of corpus

jieba.load_userdict("dict.txt")
path = "fat_news/"
corpus = []

for root, dirs, files in os.walk(path):
    # print root
    for f in files:
        doc = ""
        date = 0
        url = ""
        title = ""

        file_path = os.path.join(root, f)
        with open( file_path ) as op:
            for i,line in enumerate(op):
                if i is 0:
                    url = line
                elif i is 2:
                    date = line # use later
                elif i is 4:
                    title = line
                else:
                    doc += line
        fileread = lambda filename: open(filename, "rb").read()
        coding = chardet.detect(fileread(file_path))['encoding']
		# {'confidence': 0.99, 'encoding': 'utf-8'}
        doc = doc.decode( coding )
        
        seg_list = jieba.cut_for_search(doc)  # 搜索引擎模式
#         print(" ".join(seg_list).encode('utf-8'))
        result_list = " ".join(seg_list).encode('utf-8')
        result_list = tokenizer(stop_words,result_list)
        corpus.append(result_list)
        
# print corpus
for word in corpus:    
    print len(word)
```

```python
def _to_unicode(val):
    if val is None or isinstance(val, unicode):
        return val
    return val.decode('utf-8')
```

```python
## split words ##
texts = [[word for word in document.split()] for document in corpus]

print len(texts)
for sentence in texts:
    for word in sentence:
        print _to_unicode(word),
```

```python
## construct corpora ##
all_tokens = sum(texts, [])
tokens_once = set(word for word in set(all_tokens) if all_tokens.count(word) == 1)
texts = [[word for word in text if word not in tokens_once] for text in texts]
```

# 建字典

```python
## construct dictionary ##
dictionary = corpora.Dictionary(texts)
```

```python
print "dictionary: ",
print dictionary
```

```python
print "dictionary.token2id: "
for i,word in zip(range(len(dictionary.token2id)),dictionary.token2id):
    print i ,":", word
print len(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in texts]

tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]

for doc in corpus_tfidf:
    print len(doc)
```

# 補洞洞

```python
## add space ## 因為gensim沒有補滿
docVec = []

for doc in corpus_tfidf:
	i = 0
	data = []
	for item in range(len(dictionary.token2id)):
		if doc[i][0] == item:
			if i < len(doc)-1:
				i = i + 1
				# print "Yes",
				data.append(doc[i][1])
		else:
			# print "No",
			data.append(0)
	docVec.append(data)
        
print len(docVec)
print docVec

################### 想不通哇哇啊啊哇哇啊 ####################
# print docVec[0].index(max(docVec[0]))
print docVec[0]  
```
